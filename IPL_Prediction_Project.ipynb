```python
# Removing unwanted columns
match_df.drop(['toss_winner', 'toss_decision', 'umpire1', 'umpire2', 'umpire3', 'player_of_match'], axis=1, inplace=True)
```

```python
delivery_df.drop(['bowler', 'is_super_over', 'wide_runs','bye_runs', 'legbye_runs', 'noball_runs', 'penalty_runs','dismissal_kind', 'fielder'], axis=1, inplace=True)
```

* batsman_runs >> Runs made by batsman
* total_runs >> Runs scored (if no ball, wide, extra runs etc)

```python
# Extracting total runs data according to innings 
# inning 1 -> First team to Bat
# inning 2 -> Second team to Bat
total_score_df = (delivery_df.groupby(['match_id','inning']).sum()['total_runs'].reset_index())
```

```python
total_score_df
```

```python
# First Team data
total_score_df = total_score_df[total_score_df['inning']==1]
total_score_df
```

```python
# merging total_score_df and match_df
match_df=(match_df.merge(total_score_df[['match_id','total_runs']],left_on='id', right_on='match_id'))
```

```python
match_df
```

```python
# Teams
teams = list(match_df['team1'].unique())
```

```python
teams
```

```python
# Replacing old team names with new team names
# To make future prediction results of the model more accurate
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')

match_df['team1'] = match_df['team1'].str.replace('Gujarat Lions', 'Gujarat Titans')
match_df['team2'] = match_df['team2'].str.replace('Gujarat Lions', 'Gujarat Titans')
```

```python
match_df['team1'].unique()
```

```python
teams = ['Sunrisers Hyderabad',
 'Mumbai Indians',
 'Gujarat Titans',
 'Royal Challengers Bangalore',
 'Kolkata Knight Riders',
 'Delhi Capitals',
 'Kings XI Punjab',
 'Chennai Super Kings',
 'Rajasthan Royals']
```

```python
# Removing team names which are no more existing
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df['team1'].unique()
```

```python
match_df.shape
```

```python
match_df.head()
```

```python
# dl_applied is 0 means normal match and 1 means duckworth lewis rules are applied (stoppage of match due to rain, target runs/overs are reduced)
```

```python
match_df['dl_applied'].value_counts()
```

```python
# Removing dl_applied = 1 records 
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df.shape
```

```python
# Columns only required for further data analysis
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
match_df
```

```python
# Target scores to chase in Second inning
delivery_df = delivery_df[delivery_df['inning']  == 2]
```

```python
delivery_df.info()
```

```python
delivery_df = match_df.merge(delivery_df, on='match_id')
```

```python
delivery_df
```

```python
# Renaming Columns
# 'total_runs_x' >> 'total runs scored'     'total_runs_y' >> 'individual runs scored by per ball'

delivery_df.rename(columns={'total_runs_x':'total_runs', 'total_runs_y':'ball_score'},inplace=True)
```

```python
delivery_df
```

```python
# Adding up run scores till last ball points (cummulative sum)
delivery_df['score'] = delivery_df[['match_id','ball_score']].groupby('match_id').cumsum()['ball_score']
```

```python
delivery_df[['ball_score','score']]
```

```python
# New column which has target (getting reduced after each run)
delivery_df['target_left'] = (delivery_df['total_runs'] + 1) - delivery_df['score']
```

```python
delivery_df
```

```python
# 1 over >> 6 balls
# ipl has 20 over match >> 20 * 6 = 120 balls
delivery_df['remaining_balls'] = (120 - ((delivery_df['over'] - 1)* 6 + delivery_df['ball']))
```

```python
delivery_df
```

```python
# How many wickets are left out
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna('0')
```

```python
delivery_df
```

```python
# whenever a player is out its name get replaces with 1
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda X:X if X == '0' else '1').astype('int64')
```

```python
delivery_df['player_dismissed']
```

```python
# number of wickets taken cummulative sum
delivery_df['Wickets'] = delivery_df[['match_id','player_dismissed']].groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delivery_df
```

```python
delivery_df['Wickets'] = 10 - delivery_df['Wickets']
```

```python
delivery_df = delivery_df.rename(columns={'Wickets': 'wicket_left'})
```

```python
delivery_df
```

Current Run Rate


$$Current \ run \ rate = \frac{Runs \ scored}{Number \ of \ Overs \ completed}$$

```python
delivery_df['crr'] = (delivery_df['score'])*6/(120 - delivery_df['remaining_balls'])
```

Required Run Rate


$$Required \ run \ rate = \frac{Remaining \ score \ or \ Target}{Number \ of \ Overs \ left}$$

```python
delivery_df['rrr'] = (delivery_df['target_left'])*6/(delivery_df['remaining_balls'])
```

```python
delivery_df
```

```python
# 2nd inning data
# Creating a column result   
# where 1 >> batting_team won   0 >> batting_team lost
```

```python
def result(row):
    if row['batting_team'] == row['winner']:
        return 1
    else:
        return 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
delivery_df.columns
```

```python
Model_df = delivery_df[['batting_team','bowling_team', 'city', 'score','target_left', 'remaining_balls', 'wicket_left', 'crr', 'rrr', 'result' ]]
```

```python
# Final data before data processing 
Model_df
```

```python
Model_df.isnull().sum()
```

```python
# Removing records where city is missing
Model_df[Model_df['city'].isna()]
```

```python
Model_df.dropna(inplace=True)
```

```python
Model_df.isnull().sum()
```

```python
Model_df.describe()
```

```python
# Removing infinity values (where denominator values in rrr formulae = 0 )
Model_df = Model_df[Model_df['remaining_balls'] != 0]
```

```python
Model_df.describe()
```

```python
# shuffling the data so that no bias is created while splitting the data
```

```python
Model_df = Model_df.sample(Model_df.shape[0])
```

```python
# Random sample data is picked 
Model_df.sample()
```

```python
#Splitting the data into features and target

X = Model_df.iloc[:,:-1]
Y = Model_df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
# Splitting data into train and test data
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
# Transforming categorical features into numerical

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(drop='first'),['batting_team','bowling_team','city'])], remainder = 'passthrough')
```

```python
# Dropping first row and remaining columns will be ignored
```

```python
trf
```

```python
# building the model through logistic regression
# Top most input will be pushed into model through pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
# First transformation takes place and then the logistic regression model is built
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,Y_train)
```

```python
# making predictions data
```

```python
Y_prediction = pipe.predict(X_test)
```

```python
Y_prediction
```

```python
# model evaluation with accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_prediction)
```

```python
# Exporting the model
```

```python
# importing all model files as pickle format (compressed) to build an application through pickle library
import pickle as pkl

pkl.dump(pipe,open('model.pkl','wb') )
```
